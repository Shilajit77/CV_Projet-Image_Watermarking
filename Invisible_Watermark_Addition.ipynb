```python
import os
import pathlib
import cv2
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances,log_loss
import math
import time
import sys
import random
```

```python
img = cv2.imread("1677401105669.jpg")
text = "SAMPLE WATERMARK"
text_img = np.zeros_like(img)
cv2.putText(text_img, text, (10, img.shape[0] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, 255, 2, cv2.LINE_AA)
text_img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)
text_img = cv2.threshold(text_img, 127, 255, cv2.THRESH_BINARY)[1]
alpha = 0.1
watermarked_img = cv2.addWeighted(img, 1 - alpha, cv2.cvtColor(text_img, cv2.COLOR_GRAY2BGR), alpha, 0)
cv2.imwrite("output_image.jpg", watermarked_img)
plt.imshow(watermarked_img)
```
